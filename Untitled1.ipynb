```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X1 = np.random.normal([2,2],0.8,[200,2])
X2 = np.random.normal([4,4],0.4,[200,2])

plt.figure(figsize=(6, 6))

plt.plot(X1[:,0],X1[:,1],'b.',label='Classe 1')
plt.plot(X2[:,0],X2[:,1],'r.',label='Classe 2')
plt.legend()
plt.title('Distribuições')
plt.xlabel('x_1')
plt.ylabel('x_2')
plt.show()
```

```python
#Separar os dados 90% treinamento 10% teste
ntes = 40
n = 2

teste = np.zeros((ntes,n))

for i in range(len(teste)):
    if i < int(ntes/2):
        teste[i] = X1[i]
    else:
        teste[i] = X2[i-20]


        
inicio = int(ntes/2)
X11 = X1[inicio:,:]
X22 = X2[inicio:,:]
        
plt.figure(figsize=(6, 6))
plt.plot(X11[:,0],X11[:,1],'b.',label='Classe 1')
plt.plot(X22[:,0],X22[:,1],'r.',label='Classe 2')
plt.plot(teste[:,0],teste[:,1],'k+',label='Teste')
plt.legend()
plt.title('Distribuições')
plt.xlabel('x_1')
plt.ylabel('x_2')
plt.show()

print(len(teste))
print(len(X11))
print(len(X22))
```

```python
#Cálcula média, desvio padrão e covariância
u1 = np.mean(X11, axis=0)
u2 = np.mean(X22, axis=0)

s1 = np.std(X11, axis=0)
s2 = np.std(X22, axis=0)

c1 = covariancia(X11)
c2 = covariancia(X22)

cr1 = corelacao(c1, s1[0], s1[1])
cr2 = corelacao(c2, s2[0], s2[1])

print(u1, u2)
print(s1, s2)
print(c1, c2)
print(cr1, cr2)
```

```python
p1, p2 = evidencia(X11, X22)
z = np.zeros((len(teste),1))

for i in range(len(teste)):
    x = teste[i][0]
    y = teste[i][1]
    v1 = pdf(x, y, u1[0], u1[1], s1[0], s1[1], cr1)
    v2 = pdf(x, y, u2[0], u2[1], s2[0], s2[1], cr2)
    c = classifica(v1, v2, p1, p2)
    z[i]=c
    
plt.figure(figsize=(6, 6))
plt.plot(X11[:,0],X11[:,1],'b.',label='Classe 1')
plt.plot(X22[:,0],X22[:,1],'r.',label='Classe 2')
for i in range(len(teste)):
    if z[i]==1:
        plt.plot(teste[i,0],teste[i,1],'b+') 
    elif z[i]==-1:
        plt.plot(teste[i,0],teste[i,1],'r+')

plt.legend()
plt.title('Distribuições')
plt.xlabel('x_1')
plt.ylabel('x_2')
plt.show()
print(z)
```

```python
def evidencia(y1, y2):
    len(y1)
    len(y2)
    
    p1= len(y1)/(len(y1)+len(y2))
    p2= len(y2)/(len(y1)+len(y2))
    
    return p1, p2
```

```python
def pdf(x, y, u1, u2, s1, s2, p):
    c = -1/(2*(1-np.power(p,2)))
    
    e = c * ((np.power((x-u1),2))/np.power(s1,2) + \
             (np.power((y-u2),2))/np.power(s2,2) - \
             (2*p*(x-u1)*(y-u2))/(s1*s2))
    r = 1/(2*np.pi*s1*s2*np.sqrt(1-np.power(p,2)))*np.exp(e)
    return r
```

```python
def covariancia(y1):
    u1 = np.mean(y1, axis=0)
    a = y1[:,0] - u1[0]
    b = y1[:,1] - u1[1]
    c = np.sum(a*b)/len(y1)
    return np.abs(c)

def corelacao(c, s1, s2):
    cr = c/np.sqrt(np.power(s1,2)*np.power(s2,2))
    return cr
```

```python
def classifica(v1, v2, p1, p2):
    k = v1*p1/v2*p2
    if k >= 1:
        c = 1
    else:
        c = -1
        
        
    return c
```
