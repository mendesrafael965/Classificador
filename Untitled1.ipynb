```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X1 = np.random.normal([2,2],0.8,[200,2])
X2 = np.random.normal([4,4],0.4,[200,2])

plt.figure(figsize=(6, 6))

plt.plot(X1[:,0],X1[:,1],'bo',label='Classe 1')
plt.plot(X2[:,0],X2[:,1],'ro',label='Classe 2')
plt.legend()
plt.title('Distribuições')
plt.xlabel('x_1')
plt.ylabel('x_2')
plt.show()
```

```python
#Separar os dados 90% treinamento 10% teste
Xt = np.concatenate((X1, X2))
ntre = int(len(Xt)*0.9)
ntes = int(len(Xt)*0.1)
n = len(Xt[0])

treinamento = np.zeros((ntre,n))
teste = np.zeros((ntes,n))

#Gera indicies randomicos
ri = np.random.permutation(len(Xt))

for i in range(len(Xt)):
    if i < ntes:
        teste[i] = Xt[ri[i]]
    else:
        treinamento[i-ntes] = Xt[ri[i]]
        
plt.figure(figsize=(6, 6))
plt.plot(treinamento[:,0],treinamento[:,1],'bo',label='Treinamento')
plt.plot(teste[:,0],teste[:,1],'ro',label='Teste')
plt.legend()
plt.title('Distribuições')
plt.xlabel('x_1')
plt.ylabel('x_2')
plt.show()

print(len(teste))
print(len(treinamento))
```
